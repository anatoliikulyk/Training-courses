```python
import numpy as np
import pandas as pd
from sklearn import model_selection,  datasets, metrics, cross_validation, svm
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import SGDRegressor
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score, KFold, GridSearchCV
```

```python
newsgroups = datasets.fetch_20newsgroups(subset='all', categories=['alt.atheism', 'sci.space'])
```

```python
newsgroups.keys()
```

```python
print(newsgroups.data[-2])
```

```python
y=newsgroups.target
```

```python
print (len(newsgroups.data))
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(newsgroups.data)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()[3333]
```

```python
grid = {'C': np.power(10.0, np.arange(-5, 6))}
cv = KFold(n_splits=5, shuffle=True, random_state=241)
clf = svm.SVC(kernel='linear', random_state=241)
gs = GridSearchCV(clf, grid, scoring='accuracy', cv=cv)
gs.fit(X, y)
```

```python
gs.best_params_
```

```python
gs.best_estimator_.fit(X,y)
```

```python
Coef=gs.best_estimator_.coef_.toarray()
```

```python
Coef.shape
```

```python
l=list(zip(Coef[0], range(0,Coef.shape[1])))
l
```

```python
l.sort(key=lambda x : abs(x[0]),reverse=True)
```
